```python
import time

class Clock(object):
    
    def __init__(self, hour=0, minute=0, second=0):
        # Initialize method
        self._hour = hour
        self._minute = minute
        self._second = second
        
    def run(self):
        # Running
        self._second += 1
        if self._second == 60:
            self._second = 0
            self._minute += 1
            if self._minute == 60:
                self._minute = 0
                self._hour += 1
                if self._hour == 24:
                    self._hour =0
                    
    def show(self):
        #Show the time
        return '%02d:%02d:%02d'% (self._hour, self._minute, self._second)
    
def main():
    clock = Clock(23,59,58)
    while True:
        print(clock.show())
        time.sleep(1)
        clock.run()
        
if __name__ == '__main__':
    main()
```

```python
import math

class Point(object):
    
    def __init__(self, x=0, y=0):
        """Initialize method
        
        :param x:  X-axis
        :param y: Y-axis
        """
        self.x = x
        self.y = y
        
    def move_to(self, x, y):
        """Move to the specified location
        
        :param x: New x-coordinate
        "param y: New y-coordinate
        """
        self.x = x
        self.y = y
        
    def move_by(self, dx, dy):
        """Moves the specified increment
        
        :param dx: Increment in x-coordinate
        "param dy: Increment in y-coordinate
        """
        self.x += dx 
        self.y += dy
    
    def distance_to(self, other):
        """Calculate the distance from the other point
        
        :param other: The other point
        """
        dx = self.x - other.x
        dy = self.y - other.y
        return math.sqrt(dx ** 2 + dy ** 2)
    
    def __str__(self):
        return '(%s, %s)' % (str(self.x), str(self.y))
    
def main():
    p1 = Point(3, 5)
    p2 = Point()
    print(p1)
    print(p2)
    p2.move_by(-1, 3)
    print(p2)
    print(p1.distance_to(p2))
    
    
if __name__ == '__main__':
    main()
```
